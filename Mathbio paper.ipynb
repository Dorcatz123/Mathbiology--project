# Abstract
&nbsp;

<span style="font-size: 14pt; font-weight: normal; color: #333;">
A tumor growth model incorporating the necrotic part and using mixed order pharmacokinetics of the drug (bevacizumab) is presented in [1].
. 
</span>

&nbsp;

<span style="font-size: 14pt; font-weight: normal; color: #333;">
The model is able to show good fit to the experimentally observed results and is able to explain the important physiological phenomena. The mathematical analysis of the model is lacking. The aim of the current project is to use this model of tumor growth and do further mathematical analysis and simulations.
</span> 


&nbsp;

<span style="font-size: 14pt; font-weight: normal; color: #333;">
The experimental data is available for the treatment of andenocarcinoma tumor on laboratory mice. Two types of treatments were performed: 
    
&nbsp;
    <ol>    
      <li> a large dosage of 10 mg/Kg (0.171 mg/ml serum level) of drug at the beginning of the treatment.</li> 
      <li> Moderate dosage of 1/18 mg/Kg (9.5 $\times 10^{-4}$mg/ml serum level) for each day for 18 days.</li>
    </ol>
    
&nbsp;
    
The model is able to better fit the experimental data from continuous dosage treatment in contrast to the single dosage treatment. Further predictions are made based on the simulation results by varying the type of treatment.   
    
</span>    

## Mathematical Model

&nbsp;

<span style="font-size: 14pt; font-weight: normal; color: #333;">

 Let $X_{1}$ denotes the proliferating tumor volume $(mm^{3})$, $X_{2}$ denotes the necrotic tumor volume $(mm^{3})$ and $X_{3}$ denotes the inhibitor (bevacizumab) serum level $(mg/ml)$, $O$ denotes the compartment outside the model. The following equations describe the dynamics of tumor and the serum [1]:
 
&nbsp; 
 
<ol> 
 
 <li> $X_{1} \overset{a}\rightarrow 2X_{1}$ represents the growth of tumor cells with the rate $a$. Hence using the mass action kinetics we get $\dot{x_{1}}=ax_{1}$. </li>
    
&nbsp;  
    
 <li> $X_{1} \overset{n}\rightarrow X_{2}$ represents the proliferating tumor cells becoming necrotic tumor cells at the rate $n$. Hence by the mass action kinetics we obtain $\dot{x_{1}}=-nx_{1}$ and $\dot{x_{2}}=nx_{1}$. </li>

&nbsp;     
    
 <li> $X_{3} \overset{c}\rightarrow O$ represents the clearance rate of the drug or the outflow of the drug to a compartment outside the model at the rate $c$. By using the Michealis-Menten kinetics we get $\dot{x_{3}}=-c\frac{x_{3}}{K_{B}+x_{3}}$ where $K_{B}$ is the Michealis-Menten constant for the inhibitor. </li>

&nbsp;    
    
 <li> $X_{1} + X_{3} \overset{b}\rightarrow X_{2}$ represents the inhibitor binding to the VEGF molecules whose  concentration is proportional to the volume of tumor with the rate $b$. Hence by using the Michealis- Menten kinetics we get $\dot{x_{1}}=-bx_{1}\frac{x_{3}}{ED_{50} + x_{3}}$, $\dot{x_{2}}=bx_{1}\frac{x_{3}}{ED_{50} + x_{3} }$ and $\dot{x_{3}}=-b_{k}x_{1}\frac{x_{3}}{ED_{50} + x_{3} }$. Where $b_{k}=bk$ and $k$ is a constant with the dimension $(mg/ml)/mm^{3}day$. $ED_{50}$ is the median effective dosage which is found from [2] to be $5 \times 10^{-5}mg/ml$.
  
    
</ol>    
    
    
</span>    

<span style="font-size: 14pt; font-weight: bold; color: blue;">
Combining the above parts into differential equations will give the following models:
</span>    
    

# The differential equations can be written as follows:


\begin{equation}\label{eq1}
    \dot{x_{1}} = (a-n)x_{1} - bx_{1}\frac{x_{3}}{ED_{50}+x_{3}}.
\end{equation}

\begin{equation}\label{eq2}
    \dot{x_{2}} = nx_{1} + b\frac{x_{1}x_{3}}{ED_{50}+x_{3}}.
\end{equation}

\begin{equation}\label{eq3}
    \dot{x_{3}} = -b_{k}\frac{x_{1}x_{3}}{ED_{50}+x_{3}} - c\frac{x_{3}}{K_{B}+x_{3}} + u. \tag{1}
\end{equation}

&nbsp;  

<span style="font-size: 14pt; font-weight: normal; color: #333;">
\(u\) is a function of the input dosage rate of the inhibitor. \(u\) will vary according to the type of treatment used.
</span>

&nbsp;  


<span style="font-size: 14pt; font-weight: normal; color: #333;">
The model for the treatment of tumor using a single initial dosage is similar to the above model:
</span>


\begin{equation}\label{eq4}
    \dot{x_{1}} = (a-n)x_{1} - bx_{1}\frac{x_{3}}{ED_{50}+x_{3}}.
\end{equation}

\begin{equation}\label{eq5}
    \dot{x_{2}} = nx_{1} + b\frac{x_{1}x_{3}}{ED_{50}+x_{3}}.
\end{equation}

\begin{equation}\label{eq6}
    \dot{x_{3}} = -b_{k}\frac{x_{1}x_{3}}{ED_{50}+x_{3}} - c\frac{x_{3}}{K_{B}+x_{3}}.\tag{2}
\end{equation}

&nbsp;  


<span style="font-size: 14pt; font-weight: normal; color: #333;">
with the condition that \(x_{3}(0)\) is the initial serum drug level. The single dosage treatment that will be considered throughout the project is the treatment on mice (C1-C5) with \(x_{3}(0) = 0.171 \, \text{mg/ml}\), which is the initial serum level of the inhibitor.

</span>

# Analysis

## Theorem 1 


<span style="font-size: 14pt; font-weight: normal; color: #333;">
The system eq(1) has a unique solution and $\{(x,y,z): x\geq 0, y\geq 0, z\geq0\}$ is a positive invariant set for the system.
</span>


## **proof:**

<span style="font-size: 14pt; font-weight: normal; color: #333;">
Note that $\dot{x_{1}}|_{x_{1}=0} \geq 0$, $\dot{x_{2}}|_{x_{2}=0}=nx_{1}+b\frac{x_{1}x_{3}}{Ed_{50}+x_{3}}  \geq 0$ if $x_{1}$, $x_{3}\geq 0$ and $\dot{x_{3}}|_{x_{3}=0}=u\geq 0$. Hence it follows from the proposition 2.1 in [3] that $\{(x,y,z): x\geq 0, y\geq 0, z\geq0\}$ is a positive invariant set for the system.
The equations on the right hand side of eq(1) are $C^{1}$ functions in the domain $\{t:t\geq0\}\times \mathbb{R}^{3}_{+}$. By the existence and uniqueness theorem the system eq(1) has a unique solution in $\{t:t\geq0\}\times \mathbb{R}^{3}_{+}$.  
</span>




## Theorem 2


<span style="font-size: 14pt; font-weight: normal; color: #333;">
    
The proliferating tumor volume is bounded by exponential functions as follows
    
$$
 x_{1}(0)e^{(a-n-b)t}\leq x_{1}(t)\leq x_{1}(0)e^{(a-n)t}.
$$

If $(a-n)<0$ then eventually all the proliferating tumor will become necrotic tumors.
    
</span>


## **Proof:**

<span style="font-size: 14pt; font-weight: normal; color: #333;">
    
From eq(1) it follows that $(a-n-b)x_{1} \leq \dot{x_{1}} \leq (a-n)x_{1}.$  
Hence it follows that the solution $x_{1}(t)$ satisfies the inequality
    
</span>    

&nbsp; 
    
$$
    x_{1}(0)e^{(a-n-b)t}\leq x_{1}(t) \leq x_{1}(0)e^{(a-n)t}.
$$

&nbsp; 



<span style="font-size: 14pt; font-weight: normal; color: #333;">
Note that from eq(1) it follows that the dynamics of proliferating tumor ($x_{1}$) and the inhibitor ($x_{3}$) is independent of the dynamics of necrotic tumor ($x_{2}$). Hence we focus on the system 
</span>
    
    
$$
    \dot{x_{1}}=(a-n)x_{1}-bx_{1}\frac{x_{3}}{ED_{50}+x_{3}}. \tag{3}
$$


$$
    \dot{x_{3}}=-b_{k}\frac{x_{1}x_{3}}{ED_{50}+x_{3}}-c\frac{x_{3}}{K_{B}+x_{3}}+u. \tag{4}
$$


&nbsp; 

<span style="font-size: 14pt; font-weight: normal; color: #333;">
For all the further analysis let us assume that $u$ is a constant function. 
</span>


## Theorem 3


<span style="font-size: 14pt; font-weight: normal; color: #333;">
The system eq(3)-eq(4) has a locally asymptotically stable equilibrium point $(0,u.\frac{K_{B}}{c-u})$ if $a-n < 0$.
</span>

## **Proof:**

<span style="font-size: 14pt; font-weight: normal; color: #333;">
Solving the equations 

$$
(a-n)x_{1}-bx_{1}\frac{x_{3}}{ED_{50}+x_{3}}=0.$$
$$
-b_{k}\frac{x_{1}x_{3}}{ED_{50}+x_{3}}-c\frac{x_{3}}{K_{B}+x_{3}}+u=0.
$$




we get two equilibrium points 


$$(\bar{x_{1}},\bar{x_{3}})=(0,u.\frac{K_{B}}{c-u}), (x_{1}^{*},x_{3}^{*})=\big(\frac{b}{b_{k}.(a-n)}.(u-c.\frac{ED_{50}}{(K_{B}.\frac{b-a+n}{a-n}+ED_{50})}), \frac{(a-n)ED_{50}}{b-a+n}\big)$$.

Evaluating the Jacobian of the system eq(3)-eq(4) we get 


$$
J|_{(\bar{x_{1}},\bar{x_{3}})}=\begin{bmatrix}
 a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})} & 0\\
 \frac{-b_{k}.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})} & \frac{-c.K_{B}}{(K_{B} + \frac{u.K_{B}}{c-u})^{2}}
\end{bmatrix}  \tag{5}
$$


\begin{equation}\label{eq10}
    J|_{(x_{1}^{*},x_{3}^{*})}= \begin{bmatrix}
    0 & \frac{-b.x_{1}^{*}.ED_{50}}{(ED_{50} + x_{3}^{*})^{2}}\\
    \frac{-b_{k}.x_{3}^{*}}{(ED_{50}+x_{3}^{*})} & \frac{-b_{k}.x_{1}^{*}.ED_{50}}{(ED_{50}+x_{3}^{*})^{2}}-\frac{c.K_{B}}{(K_{B}+x_{3}^{*})^{2}}
    \end{bmatrix}
\end{equation}

</span>    
    
    
&nbsp;    
    
    
<span style="font-size: 14pt; font-weight: normal; color: #333;">   
    
Let us assume that the equilibrium point $(x_{1}^{*},x_{3}^{*})$ is in $\mathbb{R}^{2}_{+}$. It follows that  $\dot{x_{2}}|_{(x_{1}^{*},x_{3}^{*})}=x_{1}^{*}(n + b.\frac{x_{3}^{*}}{ED_{50}+x_{3}^{*}})=a.x_{1}^{*}$, since $x_{1}^{*}\neq 0$ then the necrotic tumor $x_{2}$ increases in mass for all future time, which is absurd because the proliferating tumor is at equilibrium (with the value $x_{1}^{*}$) and the serum inhibitor level is a constant (with the value $x_{3}^{*}$). 
</span>
    
    
&nbsp;      
    
<span style="font-size: 14pt; font-weight: normal; color: #333;">       
This implies that contrary to our assumption $(x_{1}^{*},x_{3}^{*})$ doesn't belong to $\mathbb{R}^{2}_{+}$. We are left with the case when $x_{1}^{*}=0$ which is possible if and only if the dosage rate is a fixed value of $u=c.\frac{ED_{50}}{(K_{B}.\frac{b-a+n}{a-n}+ED_{50})}$. At this dosage rate both the equilibrium points $(x_{1}^{*},x_{3}^{*})$ and $(\bar{x_{1}},\bar{x_{3}})$ are equal.
</span>

&nbsp;  

<span style="font-size: 14pt; font-weight: normal; color: #333;">       
For the equilibrium point $(\bar{x_{1}},\bar{x_{3}})$, let $c > u$ so that the equilibrium point is a positive value. From [3] it follows that the eigenvalues are $a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})}$ and 
    
$\frac{-c.K_{B}}{(K_{B} + \frac{u.K_{B}}{c-u})^{2}}$. If $a-n < 0$ then the eigenvalues are negative. Hence the equilibrium point $(\bar{x_{1}},\bar{x_{3}})$ is locally asymptotically stable. This equilibrium is a 'favourable' equilibrium since the proliferating tumor $\bar{x_{1}}=0$ and from eq(1) it follows that necrotic tumor is stable (i.e $\dot{x_{2}}|_{(\bar{x_{1}},\bar{x_{3}})}=0$). 
   
</span>



## Theorem 4

<span style="font-size: 14pt; font-weight: normal; color: #333;">  
If $a-n > 0$ and $K_{B}.\frac{b-a+n}{a-n}+ED_{50} > 0$ then the dosage rate $u=c.\frac{ED_{50}}{(K_{B}.\frac{b-a+n}{a-n}+ED_{50})}$ is a bifurcation value for the  system eq(3)-eq(4).
</span>


## **Proof**

<span style="font-size: 14pt; font-weight: normal; color: #333;">  
From eq(5) it follows that the eigenvalues are $a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})}$ and $\frac{-c.K_{B}}{(K_{B} + \frac{u.K_{B}}{c-u})^{2}}$. It follows that $a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})} < 0$ if and only if $u > c.\frac{ED_{50}}{(K_{B}.\frac{b-a+n}{a-n}+ED_{50})}$ and  $a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})} > 0$ if and only if $u < c.\frac{ED_{50}}{(K_{B}.\frac{b-a+n}{a-n}+ED_{50})}$. Hence if $u > c.\frac{ED_{50}}{(K_{B}.\frac{b-a+n}{a-n}+ED_{50})}$ then the equilibrium point is locally asymptotically stable (eigenvalues are negative). If $u < c.\frac{ED_{50}}{(K_{B}.\frac{b-a+n}{a-n}+ED_{50})}$ then the equilibrium point is a saddle (eigenvalues are of opposite sign).
</span>
    




<span style="font-size: 14pt; font-weight: normal; color: #333;">  
The biological interpretation of the previous theorem is as follows: If the tumor growth rate is greater than the necrosis rate (i.e $a > n$) and if the inhibition rate ($b$) is greater than the net tumor growth rate ($a-n$) then the dosage rate $u$ is a bifurcation parameter for the system eq(3)-eq(4). 
</span>

&nbsp;    

<span style="font-size: 14pt; font-weight: normal; color: #333;">  
Note that the system eq(1) has a family of equilibrium points $\{(0,x_{2},u.\frac{K_{B}}{(c-u)})\}$ for all $x_{2}$ such that $0 \leq x_{2}$. The jacobian matrix at any of these equilibrium points is given by 
</span>


$$
J|_{(\bar{x_{1}},\bar{x_{3}})}=\begin{bmatrix}
 a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})} & 0 & 0\\
 n+\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B})}{c-u}}) & 0 & 0\\
 \frac{-b_{k}.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})} & 0 & \frac{-c.K_{B}}{(K_{B} + \frac{u.K_{B}}{c-u})^{2}}
\end{bmatrix}
$$

&nbsp;   


<span style="font-size: 14pt; font-weight: normal; color: #333;"> 
The eigenvalues are clearly $a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})}$, 0 and $\frac{-c.K_{B}}{(K_{B} + \frac{u.K_{B}}{c-u})^{2}}$. Hence the stability of the equilibrium point is determined by the sign of the eigenvalue $a-n-\frac{b.u.K_{B}}{(c-u).(ED_{50} + \frac{u.K_{B}}{c-u})}$.
</span>

## Theorem 5
<span style="font-size: 14pt; font-weight: normal; color: #333;">  
There is no periodic orbit for the system eq(3)-eq(4) in the region $\{(x_{1},x_{3}): x_{3} \geq \frac{(a-n)ED_{50}}{(b-a+n)}\}$. 
</span>    

## **Proof**

<span style="font-size: 14pt; font-weight: normal; color: #333;">  
Let $f=(a-n)x_{1}-bx_{1}\frac{x_{3}}{ED_{50}+x_{3}}$ and $g=-b_{k}\frac{x_{1}x_{3}}{ED_{50}+x_{3}}-c\frac{x_{3}}{K_{B}+x_{3}}+u$. Since $\frac{\partial{f}}{\partial{x_{1}}}+\frac{\partial{g}}{\partial{x_{3}}}=(a-n)-\frac{b.x_{3}}{(ED_{50}+x_{3})}-\frac{b_{k}.x_{1}.ED_{50}}{(ED_{50}+x_{3})^{2}}-\frac{c.K_{B}}{(K_{B}+x_{3})^{2}} \leq 0$ when $(a-n)-\frac{b.x_{3}}{(ED_{50}+x_{3})} \leq 0$ or equivalently when $x_{3}\geq \frac{(a-n)ED_{50}}{(b-a+n)}$. Hence by the Dulac's criteria the system eq(3)-eq(4) has no periodic solution in the region $\{(x_{1},x_{3}): x_{3} \geq \frac{(a-n)ED_{50}}{(b-a+n)}\}$.  
</span>



## Theorem 6
<span style="font-size: 14pt; font-weight: normal; color: #333;">  
The system eq(1) has no periodic orbit in the region $\{(x_{1},x_{2},x_{3}): x_{1}\geq 0, x_{2}\geq 0, x_{3}\geq 0 \}$.
</span>

## **Proof**

<span style="font-size: 14pt; font-weight: normal; color: #333;">  
Equation 2nd equation of eq(1) implies that $x_{2}$ is an increasing function. Hence it is impossible to have a periodic orbit in the positive region. 
</span>
